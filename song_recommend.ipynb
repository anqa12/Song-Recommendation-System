## Data Analysis

### Average popularity by playlist genre

```python
data['playlist_genre'].unique()
```

```python
playlist_popularity = data.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=playlist_popularity.values, y=playlist_popularity.index, palette='viridis')

for index, value in enumerate(playlist_popularity.values):
    plt.text(value, index, f'{value:.2f}', va='center', fontsize=8)

plt.title("Average Playlist Popularity by Genre")
plt.xlabel("Average Popularity")
plt.ylabel("Playlist Genre")
plt.xticks(fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

### Average Popularity Trend of all music tracks over the years

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='track_popularity', data=data, marker='o', errorbar='sd')
plt.title("Popularity Trend by Genre Over Years")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.grid(False)
plt.show()
```

### Distribution of Numerical Variables

```python
# Select only numerical columns
numerical_columns = data.select_dtypes(include=['number'])

# Plot histograms for all numerical columns
numerical_columns.hist(bins=15, figsize=(15, 10))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
# Frequency of each genre
genre_counts = data['playlist_genre'].value_counts()
print(genre_counts)

# Top 10 most frequent artists
top_artists = data['track_artist'].value_counts().head(10)
print()
print(top_artists)
```

### Correlation Matrix

```python
corr_matrix = numerical_columns.corr()
corr_matrix.style.background_gradient(cmap='coolwarm')
```
